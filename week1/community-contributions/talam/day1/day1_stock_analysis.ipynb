```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# Load environment variables
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
```

```python

system_prompt = """ You are an expert stock analyst.
You analyze stocks and suggests or recommends potentially multibagger stocks over the next 5 years.
You analyze few of the exisiting multibagger stocks in the last 5 years , do the comparative study across the potential multibagger stocks in the next 5 years.
Consider the current geo political situations, on going wars recommend 3-5 sectors that will drive huge investment and suggest top 5 equity performers in each of the sectots.
Explain the vision in 1-3 sentences of each of the stocks , risk associated and the benefits.
Identify the fradulent or manupulative stocks and eliminate them from recommending.  
"""
user_prompt = """
    Share some penny, smallcaps and midcaps stocks in Indian stock market that trades very high volume every day.
    The stocks should have high promoter holding, less or no pledges, sufficient positive free cash flow to run the operations.
    The stock should have 3 or 5 year CAGR > 30 and good ROCE.
    Consider IPOs from stocks debut last 6-12 months also.
    You can use https://www.tickertape.in/ and https://www.screener.in/ for evaluation but don't restrict the evalution from only these websites. 
"""

# Step 2: Make the messages list

messages = [
    {"role" : "system", "content": system_prompt},
    {"role" : "user", "content": user_prompt}
] # fill this in

# Step 3: Call OpenAI
openai = OpenAI()
response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)

# Step 4: print the result
print(response.choices[0].message.content)

```
